```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np
import torch

# ================================
# LOAD MODELS
# ================================
teacher = YOLO("/content/drive/MyDrive/AI project/runs/detect/train19/weights/best.pt")
student = YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/finetune2/weights/best.pt")
studentCWD = YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student4/weights/best.pt")
studentMGD = YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student9/weights/best.pt")

model_names = [
    "Teacher Model",
    "Pruned & Fine-tuned Student",
    "CWD (T=1)",
    "MGD (a=6e-5)"
]

# ================================
# VALIDATE MODELS â†’ GET METRICS
# ================================
r_teacher = teacher.val(data="/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/data.yaml", split="test")
r_student = student.val(split="test")
r_cwd = studentCWD.val(split="test")
r_mgd = studentMGD.val(split="test")

m_teacher = r_teacher.results_dict
m_student = r_student.results_dict
m_cwd = r_cwd.results_dict
m_mgd = r_mgd.results_dict

# ================================
# EXTRACT METRICS
# ================================
def extract_metrics(m):
    precision = m["metrics/precision(B)"]
    recall    = m["metrics/recall(B)"]
    f1        = 2 * precision * recall / (precision + recall + 1e-9)
    map50     = m["metrics/mAP50(B)"]
    return [precision, recall, f1, map50]

vals_teacher = extract_metrics(m_teacher)
vals_student = extract_metrics(m_student)
vals_cwd = extract_metrics(m_cwd)
vals_mgd = extract_metrics(m_mgd)

metric_labels = ["Precision", "Recall", "F1 Score", "mAP50"]

# ================================
# PLOT METRIC COMPARISON
# ================================
x = np.arange(len(metric_labels))
width = 0.20

plt.figure(figsize=(13, 6))

colors = ["#d62728", "#1f77b4", "#ff7f0e", "#2ca02c"]

bars_teacher = plt.bar(x - 1.5*width, vals_teacher, width, label=model_names[0], color=colors[0])
bars_student = plt.bar(x - 0.5*width, vals_student, width, label=model_names[1], color=colors[1])
bars_cwd     = plt.bar(x + 0.5*width, vals_cwd,     width, label=model_names[2], color=colors[2])
bars_mgd     = plt.bar(x + 1.5*width, vals_mgd,     width, label=model_names[3], color=colors[3])

def add_labels(bars):
    for bar in bars:
        h = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            h + 0.01,
            f"{h:.3f}",
            ha="center",
            va="bottom",
            fontsize=9
        )

for bars in [bars_teacher, bars_student, bars_cwd, bars_mgd]:
    add_labels(bars)

plt.xticks(x, metric_labels, fontsize=12)
plt.ylabel("Metrics Score", fontsize=12)
plt.title(
    "Metrics Comparison across models",
    fontsize=12
)

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import numpy as np
import torch

# ================================
# LOAD MODELS
# ================================
models = {
    "CWDT Ï„=1.0": YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student4/weights/best.pt"),
    "CWDT Ï„=2.0": YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student5/weights/best.pt"),
    "CWDT Ï„=3.0": YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student6/weights/best.pt"),
    "MGD a=2e-5": YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student/weights/best.pt"),
    "MGD a=4e-5": YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student7/weights/best.pt"),
    "MGD a=6e-5": YOLO("/content/drive/MyDrive/AINEWPROJECT/yolov11_prune_distillation_v2/runs/detect/student9/weights/best.pt"),
}

model_names = list(models.keys())

# ================================
# VALIDATE MODELS â†’ GET METRICS
# ================================
metrics = {}

for name, model in models.items():
    r = model.val(split="test")
    metrics[name] = r.results_dict

# ================================
# EXTRACT METRICS (UNCHANGED)
# ================================
def extract_metrics(m):
    precision = m["metrics/precision(B)"]
    recall    = m["metrics/recall(B)"]
    f1        = 2 * precision * recall / (precision + recall + 1e-9)
    map50     = m["metrics/mAP50(B)"]
    return [precision, recall, f1, map50]

vals = [extract_metrics(m) for m in metrics.values()]

metric_labels = ["Precision", "Recall", "F1 Score", "mAP50"]

# ================================
# PLOT METRIC COMPARISON
# ================================
x = np.arange(len(metric_labels))
width = 0.12

plt.figure(figsize=(14, 6))

bars = []
for i, (name, v) in enumerate(zip(model_names, vals)):
    bars.append(
        plt.bar(
            x + (i - len(vals)/2) * width + width/2,
            v,
            width,
            label=name
        )
    )

# ================================
# ADD VALUE LABELS
# ================================
def add_labels(bars):
    for bar in bars:
        h = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            h + 0.01,
            f"{h:.3f}",
            ha="center",
            va="bottom",
            fontsize=8
        )

for b in bars:
    add_labels(b)

plt.xticks(x, metric_labels, fontsize=12)
plt.ylabel("Metrics Score", fontsize=12)
plt.title("CWDT vs MGD Student Model Comparison", fontsize=13)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left")
plt.tight_layout()
plt.show()



```
